# Метод Пияковского 

##### Функция: 
$$f(x) = sin(x)$$
$$[\frac{5π}{4},2π] $$
$$ε = 0.1$$

```python
def f(x):
    return math.sin(x)
```

##### Константа Липшеца

$$L = max(f'(x)), x ∈ [\frac{5π}{4},2π]$$
$$L = max(-cos(x)) = -cos(\frac{5π}{4}) = \frac{\sqrt{2}}{2}$$

```python
L = 1/(2 ** 0.5)
```

С ее помощью будем считать нижнюю границу 

###### Точка нижний границы 

проведем через крайние точки ( a, f(a) ) ( b, f(b) ) прямую с тангенсом угла наклона равным L

$$c = \frac{1}{2L}(L(a + b) + f(a) - f(b))$$

##### Значение нижней границы

Посчитаем значение в точке пересечения двух синих прямых


$$l(x) = \frac{1}{2} (f(a) + f(b)) - L(b - a)$$

```python
import math
a = 5/4 * math.pi
b = 2 * math.pi
upper_bound = min(f(a), f(b))
eps = 0.1
count = 0
while (True): # пока зарница между границами не будет меньшье эпсилион 
    
    c = 1/(2 * L) * (L * (a + b) + f(a) - f(b)) # Взяли точку по середине 
    
    #__считаем новые нижние границы на пересечении прямых проведенных из точек:      
    l1 = 1/2 * ((f(a) + f(c)) - L * (c - a))  # А , С
    l2 = 1/2 * ((f(c) + f(b)) - L * (b - c))  # С , B соответственно
    
    
    if l1 < l2:
        b = c
    else:
        a = c
        
    upper_bound = min(upper_bound, f(c))
    lower_bound = min(l1, l2)
    count += 3
    
    if(upper_bound - lower_bound < eps):
        break
```

```python
print('[',lower_bound, upper_bound,']', 'epoch = ', count)
```
