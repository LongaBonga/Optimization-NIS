# Метод Ньютона

$$ x ^{k + 1} = x^k - t  \frac{f'(x)}{f''(x)} $$

### Функция и ее производные 

$$ f(x) = x^4 + x $$
$$ f'(x) = 4x^3 + 1 $$
$$ f''(x) = 12x^2 $$


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x ** 4 + x
```

```python
xlist = x = np.linspace(-2, 2, 100)
ylist = [f(x) for x in xlist]

plt.title("f(x)") 
plt.xlabel("x")         
plt.ylabel("y")    
plt.grid()              
plt.plot(xlist, ylist)  
```

## Зависимость новой точки от старой

$$ x ^{k + 1} = x^k - t  \frac{f'(x)}{f''(x)} $$

$$ x ^{k + 1} = x^k -  \frac{4(x^{k})^{3} + 1}{12(x^{k})^{2}} $$

```python
def cof(x):
    return (4 * x ** 3 + 1)/(12 * x ** 2)
```

```python
def h(x):
    return x + (4 * x ** 3 + 1)/(12 * x ** 2)
```

```python
fig, ax = plt.subplots(figsize=(13, 10))
xlist  = np.linspace(-5, 5, 100)
ylist = [h(x) for x in xlist]

plt.title("Зависимость новой точки от старой") 
plt.xlabel("x")         
plt.ylabel("y")    
plt.grid()  
xline = np.linspace(-4, 4, 100)
plt.plot(xlist, ylist)  
```

#### точка, где x_k+1 = x_k$$x = \frac{-1}{2^{2/3}}$$


```python
fig, ax = plt.subplots(figsize=(13, 10))
xlist  = np.linspace(-0.7, -0.575, 100)
ylist = [h(x) for x in xlist]
xline = np.linspace(-0.7, -0.575, 100)
plt.title("Зависимость новой точки от старой") 
plt.xlabel("x")         
plt.ylabel("y")    
plt.grid()          
plt.plot(xlist, ylist, xline, xline)  
```

```python
def GD(x):
    k = 1
    x_pred = x
    x = x - cof(x)
    while(x != x_pred):
        x_pred = x
        x = x - cof(x)    
        k += 1
    return x, k
```

```python
point = -50
points = []
y = []
k = []
for i in range(1000):
    y_1, k_1 = GD(point)
    points.append(point)
    y.append(y_1)
    k.append(k_1)
    point+=00.1
```

```python
fig, ax = plt.subplots(figsize=(13, 10))

plt.title("Значения к которому сходится алгоритм из точки") 
plt.xlabel("x")         
plt.ylabel("y")    
plt.grid()              
plt.plot(points, k, points, y)  
print(y[0])
```
